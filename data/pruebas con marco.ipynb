```python
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
bd = pd.read_excel("BD_IE_cons.xlsx")
bd
```

```python
df = pd.read_excel('Marco acumulativo.xlsx')
df['Código IE']=df['Código IE'].astype(str)
df = df[~df['Código IE'].isin(["Moda","Promedio"])]
df['Código IE']=df['Código IE'].astype(int)
df
```

```python
df_var = df.merge(bd, left_on='Código IE', right_on='Codigo CFK')
df_var
```

```python
niveles_num = dict(zip(["0","1A", "1B","2A", "2B", "3","4","5"],list(range(0,8))))
```

```python
df_num = df_var.replace(niveles_num)
```

```python
sns.heatmap(df_num)
```

```python
df_num.columns
```

```python
fig = px.parallel_categories(df_num,dimensions=['Dimensión 1', 'Dimensión 2', 'Dimensión 3', 'Dimensión 4',
       'Dimensión 5', 'Dimensión 6', 'Dimensión 7', 'Dimensión 8'])

fig.show()
```
