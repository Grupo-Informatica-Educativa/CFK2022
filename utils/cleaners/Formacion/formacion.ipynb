```python
respuestas_correctas = {
    '26. Raton robot':3,
    '27. Diagrama de flujo':"El programa no funciona, debe capturar nuevamente el valor de la temperatura luego de encender el ventilador",
     '28. Variable secuela':6,
     '29. A B C':"10",
    '30. Manzana Uvas Bananas':"II y III",
    '31. Variables booleanas': "La segunda expresión es verdadera si “(a O b)” es verdadero",
    "32. Alcaldía Valeria":"Imagen 3",
    '33. Botella de Color':"La botella B debe ser verde",
    '34. Alejandra codigo bloque':"Está equivocada, el resultado es 120."
}

mariaymargarita = ['24.1 No, solo está usando computadores',
 '24.2 No, los estudiantes sólo están aprendiendo a cargar cosas en un archivo compartido y crear un gráfico',
 '24.3 Si, los estudiantes están aprendiendo sobre modelos computacionales para la predicción del clima',
 '24.4 Si, los estudiantes están aprendiendo a recolectar, organizar, y analizar datos.',
                   '25.1 No, en ningún momento de habla de computadores',
 '25.2 Si, María está dividiendo el problema en problemas más sencillos',
 '25.3 Si, María está usando un algoritmo para encontrar su almuerzo',
 '25.4 No, María solo está buscando su almuerzo']


de_acuerdo = {"Totalmente en Desacuerdo":1,"En Desacuerdo":2, "Neutro":3, "De acuerdo":4,"Totalmente de acuerdo":5,
              "De Acuerdo":4,"Totalmente de Acuerdo":5, "Totalmente Desacuerdo":1, "Totalmente en desacuerdo":1, "En desacuerdo":2}

no_sabe_conoce = {"No la conozco":0, "No sé":0}

si_no = {"SI":1, "NO":0}

col_index = [ '2. Edad', '3. Contexto IDE', '4. Género',
 '5. Cabeza de Hogar',
 '6. Estado Civil',
 '7. Horas semanales', '10. Es docente STEM',
 '11. Formación']
col_caract = ['Apellido', 'Correo Electrónico', 'Curso', 'Encuesta', 'Hora',
 'ID Asignado Por Moodle', 'Nombre', 'Tipo','1. Cédula',
 '2. Edad', '3. Contexto IDE', '4. Género',
 '5. Cabeza de Hogar',
 '6. Estado Civil',
 '7. Horas semanales',
 '8. ¿Es usted líder comunitario?', '8. ¿Es usted líder comunitario?', '9. Áreas en las que enseña', '9.1 Ciencias naturales y educación ambiental', '9.10 Otro',
 '9.2 Ciencias sociales, historia, geografía, constitución política y democracia',
 '9.3 Educación artística', '9.4 Educación ética y en valores humanos',
 '9.5 Educación física, recreación y deportes', '9.6 Educación religiosa',
 '9.7 Humanidades, lengua castellana e idiomas extranjeros', '9.8 Matemáticas',
 '9.9 Tecnología e informática', '10. Es docente STEM',
 '11. Formación']

col_autoeficacia_PC = prepost.filter(regex=r'^18.[0-9]', axis=1).columns.tolist()
col_autoeficacia_tec = prepost.filter(regex=r'^17.[0-9]', axis=1).columns.tolist()
col_sentido_com = prepost.filter(regex=r'^15.[0-9]', axis=1).columns.tolist()

col_practicas_ped = prepost.filter(regex=r'^20.[0-9]', axis=1).columns.tolist()
col_practicas_eval = prepost.filter(regex=r'^22.[0-9]', axis=1).columns.tolist()

col_conocimiento = list(respuestas_correctas.keys())
col_genero_estrategias = prepost.filter(regex=r'^12.[0-9]', axis=1).columns.tolist()
col_genero_afirmaciones = prepost.filter(regex="13.[0-9]", axis=1).columns.tolist()
col_genero_estrategias
```

```python
prepost[col_genero_estrategias]
```

```python
prepost[col_genero_estrategias+mariaymargarita] = prepost[col_genero_estrategias+mariaymargarita].replace(si_no)
prepost[col_genero_afirmaciones+col_autoeficacia_PC+col_autoeficacia_tec+col_sentido_com] = prepost[col_genero_afirmaciones+col_autoeficacia_PC+col_autoeficacia_tec+col_sentido_com].replace(de_acuerdo)
prepost[col_practicas_ped+col_practicas_eval] = prepost[col_practicas_ped+col_practicas_eval].replace(no_sabe_conoce)
prepost
```

```python
df_conocimientos = prepost[['1. Cédula','Tipo']+col_conocimiento].melt(id_vars=['1. Cédula','Tipo'], value_name='Respuesta', var_name='Pregunta')
df_conocimientos['Respuesta correcta'] = df_conocimientos['Pregunta'].replace(respuestas_correctas)
df_conocimientos['Correcta'] = 1*(df_conocimientos['Respuesta'] == df_conocimientos['Respuesta correcta'])
df_conocimientos
```

```python
df_cono_pivot = df_conocimientos.pivot_table(index=['1. Cédula','Tipo'], columns='Pregunta', values='Correcta', aggfunc='max').reset_index()
df_cono_pivot
```

```python
pp_num = prepost.drop(columns=col_conocimiento)
pp_num = pd.merge(pp_num,df_cono_pivot, on=['1. Cédula','Tipo'] )
pp_num
```

```python
pp_num = pp_num.reindex(columns=col_caract+col_autoeficacia_PC+col_autoeficacia_tec+col_practicas_eval+col_practicas_ped+col_sentido_com+col_genero_estrategias+col_genero_afirmaciones+col_conocimiento+mariaymargarita)
pp_num
```

```python
pp_num.to_excel('Formacion_inicial_2106.xlsx')
```
