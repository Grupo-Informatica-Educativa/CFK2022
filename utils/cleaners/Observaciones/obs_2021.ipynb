```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("dataset_seguimiento_observaciones.xlsx")
df_1 = df[df["¿Cuántos estudiantes llevan a cabo la actividad desconectada según las instrucciones dadas?"] != "No responde"]["¿Cuántos estudiantes llevan a cabo la actividad desconectada según las instrucciones dadas?"]
df_2 = df[df["¿Cuántos estudiantes completan la actividad desconectada?"] != "No responde"]["¿Cuántos estudiantes completan la actividad desconectada?"]
#df = df[df["¿Cuántos estudiantes completan la actividad desconectada?"] != "No responde"]
#df["¿Cuántos estudiantes llevan a cabo la actividad desconectada según las instrucciones dadas?"].value_counts()
print(len(df_1))
print(len(df_2))
```

```python
df_1.value_counts()
```

```python
df_2.value_counts()
```

```python
df_union = pd.DataFrame()
df_union["contador"] = [x for x in range(len(df_1))]
df_union["¿Cuántos estudiantes llevan a cabo la actividad desconectada según las instrucciones dadas?:"] = list(df_1)
df_union["¿Cuántos estudiantes completan la actividad desconectada?:"] = list(df_2)
df_union
```

```python
for i in range(len(df_union)):
    pregunta_1 = str("¿Cuántos estudiantes llevan a cabo la actividad desconectada según las instrucciones dadas?:")
    valor_1 = pregunta_1 + str(df_union["¿Cuántos estudiantes llevan a cabo la actividad desconectada según las instrucciones dadas?:"][i]).strip()
    
    pregunta_2 = str("¿Cuántos estudiantes completan la actividad desconectada?:")
    valor_2 = pregunta_2 + str(df_union["¿Cuántos estudiantes completan la actividad desconectada?:"][i]).strip()
    
    df_union["¿Cuántos estudiantes llevan a cabo la actividad desconectada según las instrucciones dadas?:"][i] = valor_1
    df_union["¿Cuántos estudiantes completan la actividad desconectada?:"][i] = valor_2
    

df_union = df_union.rename(columns={"¿Cuántos estudiantes llevan a cabo la actividad desconectada según las instrucciones dadas?:": "pregunta1",
                                    "¿Cuántos estudiantes completan la actividad desconectada?:": "pregunta2"})
df_union
```

```python
df_union = df_union.drop(columns=["contador"])
df_union_ohe = pd.get_dummies(df_union)
df_union_ohe
```

```python
df_union_ohe.columns = df_union_ohe.columns.str.replace("pregunta1_","")
df_union_ohe.columns = df_union_ohe.columns.str.replace("pregunta2_","")
df_union_ohe = df_union_ohe.apply(pd.to_numeric)
df_union_ohe.insert(0, "contador", [x for x in range(len(df_union_ohe))])
df_union_ohe = df_union_ohe.apply(pd.to_numeric)
df_union_ohe
```

```python
df_union_melt = pd.melt(df_union_ohe, id_vars=["contador"], value_vars=df_union_ohe.columns[1:])
```

```python
arreglo_preg = []
arreglo_res = []
for i in range(len(df_union_melt)):
    arreglo_preg.append(df_union_melt["variable"][i].split(":")[0].strip())
    arreglo_res.append(df_union_melt["variable"][i].split(":")[1].strip())

df_union_melt["variable"] = arreglo_res
df_union_melt.insert(1, "pregunta", arreglo_preg)
```

```python
df_union_melt.to_excel("preguntas_alejo.xlsx")
```
