```python
import pandas as pd
```

```python
df1= pd.read_excel('../../data/descargables/EstudiantesCFK.xlsx')
```

```python
type(df1)
```

```python
df1.columns.tolist()
```

```python
#Calcular número de niños y niñas
#Tener en cuenta que edad tiene datos sucios
pivot_est1 = pd.pivot_table(df1, index=['Sexo'], values='ID', aggfunc='nunique').reset_index() #aggfunc en función de values
pivot_est1
```

```python
#Número de estudiantes por grado
pivot_est2 = pd.pivot_table(df1, index=['Grado'], values='ID', aggfunc='nunique').reset_index()
pivot_est2
```

```python
#distribución de sexos en los grados escolares
pivot_est3 = pd.pivot_table(df1, index=['Sexo','Grado'], values='ID', aggfunc='nunique').reset_index()
pivot_est3
```

```python
#Conocer capacidad de trabajo desde casa según sector de vivienda
pivot_est4 = pd.pivot_table(df1, index=["Sector vivienda","Internet",'Uso del dispositivo móvil'], values='ID', aggfunc='nunique').reset_index()
pivot_est4
```

```python
#Niños y niñas con discapacidad impactados por el proyecto
pivot_est5 = pd.pivot_table(df1, index=['Sexo','¿Te reconoces como una persona con algún tipo de discapacidad?'], values='ID', aggfunc='nunique').reset_index()
pivot_est5
```

```python
#el indicador permitirá conocer qué tanto se está propiciando el uso de esta herramienta o si lo han utilizado antes
#Se puede observar con relación al uso del dispositivo o el acceso a internet
pivot_est6 = pd.pivot_table(df1, index=['Uso del dispositivo móvil','Conoce GreenTIC'], values='ID', aggfunc='nunique').reset_index()
pivot_est6
```
